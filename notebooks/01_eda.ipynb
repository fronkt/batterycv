# batterycv EDA
Interactive view of the manifest, class balance, brightness, and run segmentation.
Run `python scripts/build_manifest.py` first. See `scripts/eda.py` for the batch version.

```python
import sys; sys.path.insert(0, '..')
import pandas as pd, matplotlib.pyplot as plt
from batterycv.config import load_paths
from batterycv.io import build_manifest, segment_runs
from batterycv.preprocess import normalize_illumination, read_bgr
paths = load_paths()
df = segment_runs(build_manifest(paths['raw_dir']), gap_s=2.0)
df.groupby('label').agg(frames=('path','size'), runs=('run_id','nunique'))
```

```python
# look at a normalized frame
import cv2
p = df[df.label=='li_ion_mobile'].path.iloc[len(df)//2]
img = normalize_illumination(read_bgr(p))
plt.figure(figsize=(10,8)); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.axis('off');
```
